#Converting Text Data to Encoded Data

```python
from sklearn.preprocessing import LabelEncoder
label =LabelEncoder()

from pandas.core.dtypes.common import is_numeric_dtype

for x in df.columns:
  if is_numeric_dtype(df[x]):
    continue
  else:
    df[x]= label.fit_transform(df[x])
```

#Separate features and target coloms

```python
x= df.drop(['price'],axis=1).values
y= df['price'].values
```

**Cross validation :** Cross validation is a technique used in machine learning to evaluate the performance of a model on unseen data.

```python
from sklearn.linear_model import LogisticRegression
reg= LogisticRegression()

#For Regression Problem
from sklearn.model_selection import cross_val_score
cross_val_score(reg, x, y, cv=2, scoring='r2')
```

Split the data into train data and test data

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2,random_state=1) #cv=1
```

```python
x.shape,x_train.shape,x_test.shape
```

#Model Implementation

```python
from sklearn.linear_model import LogisticRegression
reg= LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameter= {'penalty':['l1','l2'],
            'C':[1,2,3,4,5,6,8,10,20,30,40],
            'max_iter':[100,200,300]}
```

```python
model= GridSearchCV(reg,param_grid=parameter,scoring='r2',cv=2)
```

```python
model.fit(x_train,y_train)
```

```python
print(model.best_params_)
```

```python
print(model.best_score_)
```

```python
y_test_pred= model.predict(x_test)
```

#Model Evaluation

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,y_test_pred)
print('test part r2_score : ',score)
```

#Building a prediction system for evaluating

```python
input =[x[2]]
print(f'actual car price :[{y_train[2]}]')

predicted_class =model.predict(input)
print('predicted car price :',predicted_class)
```
